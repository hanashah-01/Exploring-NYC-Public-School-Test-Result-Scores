Every year, American high school students take SATs, which are standardized tests intended to measure literacy, numeracy, and writing skills. There are three sections - reading, math, and writing, each with a **maximum score of 800 points**. These tests are extremely important for students and colleges, as they play a pivotal role in the admissions process.

Analyzing the performance of schools is important for a variety of stakeholders, including policy and education professionals, researchers, government, and even parents considering which school their children should attend. 

You have been provided with a dataset called `schools.csv`, which is previewed below.

You have been tasked with answering three key questions about New York City (NYC) public school SAT performance.

```python
# Re-run this cell 
import pandas as pd

# Read in the data
schools = pd.read_csv("schools.csv")

# Preview the data
schools.head()
```

```python
# Display number of rows
len(schools)
```

```python
# Check for null values
schools.isna().sum()
```

```python
# Print rows with null values
schools[schools["percent_tested"].isnull()]
```

```python
# Subset columns school_name and average_math
best_math_schools = schools[["school_name","average_math"]]
```

```python
# Subset rows where average math score is at least 640
best_math_schools = best_math_schools[best_math_schools["average_math"]>=640]
```

```python
# Sort values in descending order
best_math_schools = best_math_schools.sort_values('average_math',ascending=False)
best_math_schools
```

Stuyvesant High School has the best math results, with the average math score of 754.

```python
# Add new column for total SAT scores
schools["total_SAT"] = schools["average_math"] + schools["average_reading"] + schools["average_writing"]
```

```python
# Save results in dataframe
top_10_schools = schools[["school_name","total_SAT"]]

top_10_schools = top_10_schools.sort_values("total_SAT", ascending=False).head(10)
top_10_schools
```

The top 10 performing schools are as below :-
1.  Stuyvesant High School
2.  Bronx High School of Science
3.  Staten Island Technical High School
4.  High School of American Studies at Lehman College
5.  Townsend Harris High School
6.  Queens High School for the Sciences at York College
7.  Bard High School Early College
8.  Brooklyn Technical High School
9.  Eleanor Roosevelt High School
10. High School for Mathematics, Science, and Engineering at City College

```python
# Average and standard deviation of total SAT scores for each borough
std_mean_school = schools.groupby("borough")["total_SAT"].agg(['count', 'mean', 'std']).round(2)
std_mean_school
```

```python
# Select the borough with the largest standard deviation
largest_std_dev = std_mean_school[std_mean_school["std"] == std_mean_school["std"].max()]
largest_std_dev
```

```python
# Rename the columns in the dataframe
largest_std_dev = largest_std_dev.rename(columns={"count":"num_schools", "mean":"average_SAT", "std":"std_SAT"})
largest_std_dev
```

Manhattan has the largest standard deviation in the combined SAT score.
